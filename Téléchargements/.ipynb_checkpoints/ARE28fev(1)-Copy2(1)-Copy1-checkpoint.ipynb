```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class voiture:
    
    def __init__(self, x, v, dt, proba,name):
        self.x = x
        #self.y = 2
        self.v = v
        self.dt = dt
        self.temps = dt
        self.proba = proba
        self.etat = 0 #si different de 0 alors accident et stop traffic
        self.j = name
        self.vitesse = v
        self.change = 0

    def roule(self):
        self.temps += 1
        if self.etat == 0:
            self.v = self.vitesse
            self.x = self.x + self.v*self.dt
            if self.proba == 0 :
                self.accident = 0
            else :
                self.accident = randint(1, self.proba)
            if self.accident == 1 and self.x>0:
                self.etat = 4 
                self.v = 0
        else :
            self.etat += -1
        
```

```python
class route:
    
    def __init__(self, debit, distance, x, v, dt, proba):
        #self.voitures = [voiture(x-2*i, v, dt, proba) for i in range(0,debit)]
        self.voitures1 = [voiture(0,v,dt,proba,0)]
        self.voitures2 = [voiture(0,v,dt,proba,0)]
        self.debit = debit
        self.v = v
        self.dt = dt
        self.sortie = 8
        #fin de l'ajout
        
        self.grid = np.zeros(shape=(distance,5))
        self.distance = distance
        self.state_route = self.grid
        self.proba = proba
        self.v = v
        self.dt = dt
        
        #truc pour le temps
        self.inc = 0
        self.voitures_passe = 0

    def update_route(self):
        self.state_route = np.zeros((self.distance, 5))
        #self.state_route[self.sortie][4] = 1
        #les voitures avancent
        for item in self.voitures1:
            if item.x>=0 and item.x<(self.distance-1) and self.state_route[item.x+1][1] != 0:
                item.etat = 1
                item.temps += 1
                self.state_route[item.x][1] = 5
            if item.change == 1:
                self.state_route[item.x][3] = 9
                self.voitures2.append(item)
                del item
            else :
                item.roule()
                if item.x >= 0 and item.x < self.distance and item.change ==1:
                    self.state_route[item.x][1] = 13
                    if item.etat != 0 :
                        self.state_route[item.x][1] = 5
                else:
                    if item.x >= 0 and item.x < self.distance:
                        self.state_route[item.x][1] = 9
                        if item.etat != 0 :
                            self.state_route[item.x][1] = 5
                        
        for item in self.voitures2:
            if item.x>=0 and item.x<(self.distance-1) and self.state_route[item.x+1][3] != 0:
                item.etat = 1
                item.temps += 1
                self.state_route[item.x][3] = 5
            if item.change == 1:
                self.state_route[item.x][1] = 13
                self.voitures1.append(item)
                del item
            else :
                item.roule()
                if item.x >= 0 and item.x < self.distance and item.change == 1:
                    self.state_route[item.x][3] = 9
                    if item.etat != 0 :
                        self.state_route[item.x][3] = 5
                else :
                    if item.x >= 0 and item.x < self.distance:
                        self.state_route[item.x][3] = 13
                        if item.etat != 0 :
                            self.state_route[item.x][3] = 5
                        
        
            #print([item.x for item in self.voitures])
        
        #generation de nouvelles voitures
        self.inc += 1
        if self.inc%self.debit == 0 and self.state_route[0][1] == 0 and self.state_route[1][1] == 0:
            self.voitures1.append(voiture(0,self.v,self.dt,self.proba,self.inc))
            print("plop ! une voiture est entrée dans l'autoroute !!")
        if self.inc%self.debit == 0 and self.state_route[0][3] == 0 and self.state_route[1][3] == 0:
            self.voitures2.append(voiture(0,self.v,self.dt,self.proba,self.inc))
            print("plop ! une voiture est entrée dans l'autoroute !!")
            
        #compteur de voitures passées
        #for i in self.voitures1:
            #if i.x == self.distance :
                #print( "ping ! la voiture ",i.j," a traversee l'autoroute ! elle a mis",i.temps)
                #del i
        
```

```python
road = route(1,15,0,1,1,25)
```

```python
for j in range(50):
    road.update_route()
    road.state_route
    
    plt.imshow(road.state_route)
    plt.show()
    plt.pause(0.1)
    plt.cla()
```
