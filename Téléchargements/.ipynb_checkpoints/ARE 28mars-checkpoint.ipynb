```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class voiture:
    
    def __init__(self, x, proba, name, couleur): # position de depart; proba d'accident ; nom ; couleur
        self.x = x #la position de la voiture
        self.temps = 0 # le temps passer sur la route par la voiture en tour
        self.proba = proba #proba d'accident = 1/proba
        self.etat = 0 #si different de 0 alors accident et stop traffic
        self.j = name #pour differencier les voitures
        self.couleur = couleur #pour identifier la voie d'origine

    #la fonction qui fait rouler les voitures
    def roule(self):
        #sert à connaitre le temps mis par une voiture pour traverser la route
        self.temps += 1
        if self.etat == 0: # si la voiture n'est pas accidentée
            self.x += 1 #la voiture avance
            if self.proba == 0 :
                self.accident = 0 #si on met proba = 0 jamais d'accident
            else :
                self.accident = randint(1, self.proba) #des de l'accident echec critique
            if self.accident == 1 and self.x>0:
                self.etat = 5 #tours avant de repartir
        else :
            self.etat += -1 #on enleve 1 à l'etat jusqu'a ce que la voiture puisse repartir
        
```

```python
class route:
    
    def __init__(self, debit, distance, proba): #frequence d'apparition des voitures ; longueure de la route ; accident
        #self.voitures = [voiture(x-2*i, proba) for i in range(0,debit)]
        self.voitures1 = [voiture(0,proba,0,9)] #voitures de la voie 1 à gauche
        self.voitures2 = [voiture(0,proba,0,13)] #voitures de la voie 2 à droite
        self.debit = debit
        self.sortie = 8
        #pour plus tard
        # self.grid = np.zeros(shape=(distance,5)) # inutile ???
        self.distance = distance
        self.state_route = np.zeros((self.distance, 5)) #creation de la route
        self.proba = proba
        
        self.inc = 0 # va servir a compter le temps de la route / nomer les voitures
        self.voitures_passe = 0 #compteur de voitures qui ont traversé la route

    #la fonction qui fait que le temps avance , appelee une fois par tour
    
    def update_route(self):
        self.state_route = np.zeros((self.distance, 5))
        #self.state_route[self.sortie][4] = 1
        #les voitures avancent
        
        for idx, item in enumerate(self.voitures1):#voie gauche d'abord
            #si devant la voiture il y a qqchose et que sur la voie d'à coté il n'y a aucune voiture à la meme position sans rique de colision
            if item.x>=0 and item.x<(self.distance-1) and self.state_route[item.x+1][1] != 0 and self.state_route[item.x][3] == 0 and self.state_route[item.x-1][3] == 0 and self.state_route[item.x+1][3] == 0:
                self.state_route[item.x][3] = item.couleur #la voie de droite, la voiture à changer de route
                #dans la liste de droite
                for i in range(len(self.voitures2)) :
                    #la premiere voiture dont la position est inferieure à celle de la voiture qui change de route
                    if self.voitures2[i].x < item.x :
                        self.voitures2.insert(i, item) #pour garder la liste triee par rapport à leur position x
                        print(item.x)
                        del self.voitures1[idx] #on supprime l'item dans la liste de la voie 1 puisqu'il à changer de voie
                        break # on sort
            #la voiture est sur la route et il y a un accident devant elle et elle ne peut pas changer de voie
            elif item.x>=0 and item.x<(self.distance-1) and self.state_route[item.x+1][1] != 0:
                item.etat = 2 #on force la voiture à s'arreter
                item.roule() #la voiture effectue son tour
                self.state_route[item.x][1] = 5 # on marque la position de la voiture comme etant à l'arret
            else : #RAS
                item.roule() #la voiture roule et avance
                if item.x >= 0 and item.x < self.distance:
                        self.state_route[item.x][1] = item.couleur#couleur gauche
                        if item.etat != 0 :
                            self.state_route[item.x][1] = 5#couleur arret
               
        for idx, item in enumerate(self.voitures2):
            
            if item.x>=0 and item.x<(self.distance-1) and self.state_route[item.x+1][3] != 0 and self.state_route[item.x][1] == 0 and self.state_route[item.x-1][1] == 0 and self.state_route[item.x+1][1] == 0:
                self.state_route[item.x][1] = item.couleur
                self.state_route[item.x][3] = 0
                item.etat = 0
                for i in range(len(self.voitures1)) :
                    if self.voitures1[i].x < item.x :
                        self.voitures1.insert(i,item)
                        print(item.x)
                        del self.voitures2[idx]
                        break
            elif item.x>=0 and item.x<(self.distance-1) and self.state_route[item.x+1][3] != 0:
                item.etat = 2
                item.roule()
                self.state_route[item.x][3] = 5#arret
        
            else :
                item.roule()
                if item.x >= 0 and item.x < self.distance:
                        self.state_route[item.x][3] = item.couleur
                        if item.etat != 0 :
                            self.state_route[item.x][3] = 5
                 
        
            #print([item.x for item in self.voitures])
        
        #generation de nouvelles voitures
        self.inc += 1
        if self.inc%4 == 0 and self.state_route[0][1] == 0 and self.state_route[1][1] == 0:
            self.voitures1.append(voiture(0,self.proba,self.inc,9))
            print("plop ! une voiture est entrée dans l'autoroute !!")
        if self.inc%self.debit == 0 and self.state_route[0][3] == 0 and self.state_route[1][3] == 0:
            self.voitures2.append(voiture(0,self.proba,self.inc,13))
            print("plop ! une voiture est entrée dans l'autoroute !!")
            
        
        
        #compteur de voitures passées
        #for i in self.voitures1:
            #if i.x == self.distance :
                #print( "ping ! la voiture ",i.j," a traversee l'autoroute ! elle a mis",i.temps)
                #del i
        
```

```python
road = route(2,15,25) #frequence d'apparition des voitures ; longueure de la route ; accident
```

```python
for j in range(50):
    road.update_route()
    road.state_route
    
    plt.imshow(road.state_route)
    plt.show()
    plt.pause(0.1)
    plt.cla()
```
